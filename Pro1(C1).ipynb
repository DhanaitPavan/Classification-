# **2)By Decision Tree**

```python
inputs=df1.drop('Outcome',axis='columns')
```

```python
target = df1['Outcome']
```

```python
from sklearn.preprocessing import LabelEncoder
le_pregnancies = LabelEncoder()
le_glucose = LabelEncoder()
le_bloodPressure = LabelEncoder()
le_insulin = LabelEncoder()
le_diabetesPedigreeFunction = LabelEncoder()
```

```python
inputs['pregnancies'] = le_pregnancies.fit_transform(inputs['Pregnancies'])
inputs['glucose'] = le_glucose.fit_transform(inputs['Glucose'])
inputs['bloodPressure'] = le_bloodPressure.fit_transform(inputs['BloodPressure'])
inputs['insulin'] = le_insulin.fit_transform(inputs['Insulin'])
inputs['diabetesPedigreeFunction'] = le_diabetesPedigreeFunction.fit_transform(inputs['DiabetesPedigreeFunction'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Pregnancies','Glucose','BloodPressure','Insulin','DiabetesPedigreeFunction','SkinThickness','BMI','Age'],axis='columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[6,57,16,9,70]])
```

If pregnancies=6	glucose=57	bloodPressure=16	insulin=9	diabetesPedigreeFunction=70

```python
model.predict([[1,11,13,0,44]])
```

If pregnancies=1 glucose=11 bloodPressure=13 insulin=0 diabetesPedigreeFunction=44

# **3)By KNN**

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dfa = pd.DataFrame(iris.data,columns=iris.feature_names)
dfa.head()
```

```python
dfa['target'] = iris.target
dfa.head()
```

```python
dfa[dfa.target==2].head()
```

```python
dfa['flower_name'] =dfa.target.apply(lambda x: iris.target_names[x])
dfa.head()
```

```python
dfa[45:55]
```

```python
dfx = dfa[:50]
dfy = dfa[50:100]
dfz = dfa[100:]
```

**Sepal length vs Sepal Width (Setosa vs Versicolor)**

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(dfx['sepal length (cm)'], dfx['sepal width (cm)'],color="green",marker='+')
plt.scatter(dfy['sepal length (cm)'], dfy['sepal width (cm)'],color="blue",marker='.')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(dfx['petal length (cm)'], dfx['petal width (cm)'],color="green",marker='+')
plt.scatter(dfy['petal length (cm)'], dfy['petal width (cm)'],color="blue",marker='.')
```

```python
from sklearn.model_selection import train_test_split
x = dfa.drop(['target','flower_name'], axis='columns')
y = dfa.target
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
len(X_train1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train1, y_train1)
```

```python
knn.score(X_test1, y_test1)
```

```python
knn.predict([[4.8,3.0,1.5,0.3]])
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test1)
cm1 = confusion_matrix(y_test1, y_pred)
cm1
```

```python
plt.figure(figsize=(7,5))
sn.heatmap(cm1, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test1, y_pred))
```
